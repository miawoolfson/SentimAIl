#Pre-Processing

```python
df_labeled[['cat_1_level_1', 'cat_1_level_2', 'cat_1_weight', 'labeled']].head(10)
```

```python
# Initialize empty columns
df_labeled['coarse_genre'] = None
df_labeled['included_forwarded'] = None
df_labeled['primary_topic'] = None
df_labeled['emotional_tone'] = None

# Go over cat_1 to cat_12
for i in range(1, 13):
    level_1 = f'cat_{i}_level_1'
    level_2 = f'cat_{i}_level_2'

    if level_1 in df_labeled.columns and level_2 in df_labeled.columns:
        df_labeled['coarse_genre'] = df_labeled.apply(
            lambda row: row[level_2] if row[level_1] == 1 else row['coarse_genre'], axis=1)
        df_labeled['included_forwarded'] = df_labeled.apply(
            lambda row: row[level_2] if row[level_1] == 2 else row['included_forwarded'], axis=1)
        df_labeled['primary_topic'] = df_labeled.apply(
            lambda row: row[level_2] if row[level_1] == 3 else row['primary_topic'], axis=1)
        df_labeled['emotional_tone'] = df_labeled.apply(
            lambda row: row[level_2] if row[level_1] == 4 else row['emotional_tone'], axis=1)
```

```python
df_labeled.head()
```

```python
df_labeled = df_labeled[df_labeled['cat_1_level_1'] == True].copy()
df_labeled.shape
```

```python
df_labeled['cat_1_level_1'].unique()
```

```python
coarse_genre_map = {
    1.0: "Company Business / Strategy",
    2.0: "Purely Personal",
    3.0: "Personal in Work Context",
    4.0: "Logistics",
    5.0: "Employment",
    6.0: "Document Collaboration",
    7.0: "Empty (attachment issue)",
    8.0: "Empty Message"
}
```

```python
included_forwarded_map = {
    1.0: "New Text + Forwarded Material",
    2.0: "Forwarded Emails + Replies",
    3.0: "Business Documents",
    4.0: "News Articles",
    5.0: "Gov/Academic Reports",
    6.0: "Gov Actions",
    7.0: "Press Releases",
    8.0: "Legal Documents",
    9.0: "URLs",
    10.0: "Newsletters",
    11.0: "Business Humor",
    12.0: "Unrelated Humor",
    13.0: "Attachments (missing)"
}
```

```python
primary_topic_map = {
    1.0: "Regulations",
    2.0: "Internal Projects",
    3.0: "Company Image (Current)",
    4.0: "Company Image (Influencing)",
    5.0: "Political Influence",
    6.0: "California Crisis",
    7.0: "Internal Policy",
    8.0: "Internal Operations",
    9.0: "Partnerships",
    10.0: "Legal Advice",
    11.0: "Talking Points",
    12.0: "Meeting Minutes",
    13.0: "Trip Reports"
}
```

```python
emotional_tone_map = {
    1.0: "Jubilation",
    2.0: "Hope / Anticipation",
    3.0: "Humor",
    4.0: "Camaraderie",
    5.0: "Admiration",
    6.0: "Gratitude",
    7.0: "Friendship",
    8.0: "Support",
    9.0: "Sarcasm",
    10.0: "Confidentiality",
    11.0: "Anxiety",
    12.0: "Concern",
    13.0: "Aggressiveness",
    14.0: "Triumph",
    15.0: "Pride",
    16.0: "Anger",
    17.0: "Sadness",
    18.0: "Shame",
    19.0: "Dislike / Scorn"
}
```

```python
df_labeled['coarse_genre'] = df_labeled['coarse_genre'].map(coarse_genre_map)
df_labeled['included_forwarded'] = df_labeled['included_forwarded'].map(included_forwarded_map)
df_labeled['primary_topic'] = df_labeled['primary_topic'].map(primary_topic_map)
df_labeled['emotional_tone'] = df_labeled['emotional_tone'].map(emotional_tone_map)
```

```python
df_labeled = df_labeled.drop(columns=[col for col in df_labeled.columns if col.startswith('cat_')])
```
